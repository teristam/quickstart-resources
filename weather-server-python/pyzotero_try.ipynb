```python
from pyzotero import zotero
import tempfile
```

```python
zot = zotero.Zotero(library_id='0',library_type='user',local=True)
```

```python
items = zot.top(limit=5)
for item in items:
    print(item)
```

```python
item = zot.item('877LAUK2')
item
```

```python

file = zot.file('877LAUK2')

tmp_pdf = tempfile.NamedTemporaryFile(suffix='.pdf')
tmp_pdf.name
tmp_pdf.write(file)
tmp_pdf.name
```

```python
from markitdown import MarkItDown


md = MarkItDown()
result = md.convert(tmp_pdf.name)
```

```python
with open('output.md','w') as f:
    f.write(result.text_content)
```

```python
from IPython.display import Markdown

Markdown(result.text_content)
```

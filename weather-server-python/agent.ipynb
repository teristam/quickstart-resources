```python
from smolagents import CodeAgent, OpenAIServerModel, DuckDuckGoSearchTool
import os 
from dotenv import load_dotenv

load_dotenv()

search_tool = DuckDuckGoSearchTool()

model = OpenAIServerModel('o3-mini', api_key=os.environ['OPENAI_KEY'])
agent = CodeAgent(model=model, tools=[search_tool])
```

```python
from smolagents.tools import ToolCollection
from mcp import StdioServerParameters

params = StdioServerParameters(
    command='uv',
    args = ['run',
            'pubmedmcp'],
    env={"UV_PYTHON": "3.12", **os.environ},
)

with ToolCollection.from_mcp(params) as tool_collection:
    agent = CodeAgent(model=model, tools=[*tool_collection.tools])
    agent.run('''Find a remedy for hangover. 
    Only based on results from pubmed. 
    Summarize the results and list the relevant pubmed id of the source article''')
```
